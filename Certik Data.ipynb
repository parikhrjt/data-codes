```python
def Find_Pattern(Text, Pattern):
    numOfPattern = 0
    for each in range(0, len(Text)-len(Pattern)+1):
       if Text[each:each+len(Pattern)] == Pattern:
           numOfPattern += 1
    return numOfPattern
```
